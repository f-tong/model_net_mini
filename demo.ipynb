Calculate BC

```python
import pickle
import networkx as nx
import pandas as pd

with open("./output/model_tables_bipartite/bipartite/bipartite.graph", "rb") as file:
    G = pickle.load(file)
    print("successfully loaded graph")
print("computing BC estimates")
bc = nx.betweenness_centrality(G, k = 1000)
print("successfully computed BC estimates")

with open("./output/approx_betweenness.pkl", "wb") as file:
    print("saving dict")
    pickle.dump(bc, file)
    print("successfully saved dict")
        
```

Load dataframe

```python
import pandas as pd
import pickle
with open("./output/approx_betweenness.pkl", "rb") as file:
    dict = pickle.load(file)

df = pd.DataFrame.from_dict(dict, orient="index", columns=["bc"]).reset_index().rename(columns={'index': 'homographs'})
pd.set_option('display.max_rows', None)

exclude = [
    r"\s*-\s*",
    r"[0-9]+.?[0-9]*",
]
combined_regex = '|'.join(exclude)
filtered_df = df[~df["homographs"].str.contains(combined_regex, regex=True, na=False)].sort_values(by="bc", ascending=False)

print(filtered_df.head(100))
```
